```python
ones=np.ones([x.shape[0],1])
X=np.concatenate((ones,x),axis=1)

w = np.array([0.0, 0.0, 0.0])
```

```python
def hypothesis(X, w):
    return 1/(1+np.exp(-np.dot(X, w)))
```

```python
def loss(X,y,w,m):
    return (-(1/m)*np.sum(y*np.log(hypothesis(X,w)) + (1-y)*np.log(1-hypothesis(X, w))))
```

```python
def costfun():
    pass
```

```python
def gradient(X, y, w, m):
    return (1/m)*np.dot(X.T, ((hypothesis(X,w))- y))
```

```python
m = X.shape[0]
gradient(X, y, w, m)
```

```python
# heisian matrix
def H(X, w):
    h = hypothesis(X, w)
    S = np.diag(h * (1 - h))
    hessian = (1 / X.shape[0]) * X.T.dot(S).dot(X)
    return hessian
```

```python
H(X,w)
```

```python
costVal=[]
epochs=[]
lr=0.001

for i in range(1, 3000):
    l=loss(X,y,w,m)
    costVal.append(l)
    w = w - lr * gradient(X, y, w, m)
    epochs.append(i)
        
    if i % 1 == 0:
        print(f'epoch {i}: loss = {l:.3f}')
```

```python
plt.plot(epochs, costVal)
plt.title("Cost Function vs. Epochs")
plt.xlabel("Epochs")
plt.ylabel("Cost")
plt.show()
```

### Hessian 

```python
costVal = []
epochs = []

for i in range(1, 3000):
    l = loss(X, y, w, m)
    costVal.append(l)
    grad = gradient(X, y, w, m)
    hess = H(X, w)
    w = w - lr * np.linalg.inv(hess).dot(grad)  # Update w using the Hessian matrix
    epochs.append(i)
        
    if i % 100 == 0:
        print(f'epoch {i}: loss = {l:.3f}')
```

```python
plt.plot(epochs, costVal)
plt.title("Cost Function vs. Epochs")
plt.xlabel("Epochs")
plt.ylabel("Cost")
plt.show()
```
