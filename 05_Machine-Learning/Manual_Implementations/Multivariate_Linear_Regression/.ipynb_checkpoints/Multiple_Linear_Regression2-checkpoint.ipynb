```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.loadtxt('./ex3Data/ex3x.dat', dtype=np.float32)
y = np.loadtxt('./ex3Data/ex3y.dat', dtype=np.float32)
```

```python
X = np.ones((x.shape[0], 3))
X[:, 1] = X[:, 1] * x[:, 0]
X[:, 2] = X[:, 2] * x[:, 1]
```

```python
# Scale the features
X[:, 1] = X[:, 1] * (X[:, 0] / x[:, 0].max())
X[:, 2] = X[:, 2] * (X[:, 1] / x[:, 1].max())
```

```python
def hypothesis(x, w):
    return (X.dot(w))
```

```python
def costfunction(x, w, y):
    return (1/(2 * X.shape[0])) * sum((hypothesis(X,w) - y)**2)
```

```python
def gradient(x, w, y):
    return (1/X.shape[0] * X.T.dot(hypothesis(X,w) - y))
```

```python
epochs = 300
lr = np.array([1, 0.1, 0.01, 0.001, 0.0001])
cs = ['--b', '-r', '--m', '+c', '.k']
loss = []
w = np.array([0.0, 0.0, 0.0])
for l in lr:
    loss = []
    w = np.array([0.0, 0.0, 0.0])
    for epoch in range(0, epochs):
#         epochs.append(epoch)
        loss.append(costfunction(X,w,y))
        w = w - l * gradient(X, w, y)
    plt.plot(range(epochs), loss, cs, label=str(l))
plt.legend()
plt.show()
```
