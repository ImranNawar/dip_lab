## Regularized Logistic Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.loadtxt('./ex5Data/ex5Logx.dat', delimiter=',', dtype=np.float32)
y = np.loadtxt('./ex5Data/ex5Logy.dat', dtype=np.float32)
```

```python
ones=np.ones([x.shape[0],1])
X=np.concatenate((ones,x),axis=1)

w = np.array([0.0, 0.0, 0.0])
```

```python
X.shape
```

```python
def hypothesis(X, w):
    return 1/(1+np.exp(-np.dot(X, w)))
```

```python
def cost(X, y, w, lam):
    m = X.shape[0]
    reg_term = (lam / (2 * m)) * np.sum(w[1:]**2)
    return (-(1/m)*np.sum(y*np.log(hypothesis(X,w)) + (1-y)*np.log(1-hypothesis(X, w)))) + reg_term
```

```python
w[1:]
```

```python
def gradient(X,y, w, lam):
    m = X.shape[0] # len(X)
    reg_term_g = (lam/m)*w
    reg_term_g[0] = 0
    return (1/m)*np.dot(X.T, (hypothesis(X, w) - y)) + reg_term_g
```

```python
(1/X.shape[0])*np.dot(X.T, ((hypothesis(X,w))- y))
```

```python
gradient(X, y, w, lam)
```

## For Lambda = 0

```python
costVal=[]
epochs=[]
lr=0.1
lam = 0
m = X.shape[0]

w = np.zeros(X.shape[1])

for i in range(1, 3000):
    l=cost(X, y, w, lam)
    costVal.append(l)
    w = w - lr * gradient(X,y,w,lam)
    epochs.append(i)
        
    if i % 1 == 0:
        print(f'epoch {i}: loss = {l:.3f}')
```

```python
plt.plot(epochs, costVal)
plt.title("Cost Function vs. Epochs")
plt.xlabel("Epochs")
plt.ylabel("Cost")
plt.show()
```

```python
H = hypothesis(X, w)
H
```

```python
plt.scatter(x[:,0], x[:,1], marker='o', c='r')
plt.plot(x, H, marker='+')
plt.show()
```
