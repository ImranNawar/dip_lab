```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.loadtxt('ex2Data/ex2x.dat', dtype=np.float32)
y = np.loadtxt('ex2Data/ex2y.dat', dtype=np.float32)
```

```python
X = np.ones((x.shape[0], 2))
X[:, 1] = x
```

```python
def hypothesis(x, w):
    return (np.dot(X, w))
```

```python
def cost(x, y, w):
    return (1 / (2 * X.shape[0]) * sum(hypothesis(X, w) - y)**2)
```

```python
def gradient(x, y, w):
    return (1 / X.shape[0] * X.T.dot(hypothesis(X, w) - y))
```

```python
w = np.array([0,0])
epochs = []
costVal = []

# Gradient descent
for i in range(1, 300):
    costVal.append(cost(X,y,w))
    w = w - 0.0004 * gradient(X,y,w)
    epochs.append(i)
```

```python
plt.plot(epochs, costVal)
plt.title("Cost Function vs. Epochs")
plt.xlabel("Epochs")
plt.ylabel("Cost")
plt.show()
```
