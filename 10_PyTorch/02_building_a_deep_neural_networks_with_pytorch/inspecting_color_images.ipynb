```python
import wget
url = 'https://www.dropbox.com/s/lpw10qawsc5ipbn/MyImage.JPG?dl=1'
filename = 'Image.jpg'

wget.download(url, filename)
```

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('Image.jpg')
```

```python
img = img[50:250, 40: 240, :]
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
print(img.shape)
```
