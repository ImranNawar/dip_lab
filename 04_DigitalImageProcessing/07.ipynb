```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(image, title=None):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.axis('off')
    plt.title(title)
    plt.show()
```

## Digital Image Watermarking

```python
def add_watermark(img, watermark_text):

    # Choose a font and size for the watermark
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 4
    font_thickness = 2
    font_color = (255, 255, 255)

    # Calculate the position to place the watermark
    text_size = cv2.getTextSize(watermark_text, font, font_scale, font_thickness)[0]
    text_x = img.shape[1] - text_size[0] - 150
    text_y = img.shape[0] - 50

    # Add the watermark to the image
    watermarked_img = img.copy()
    cv2.putText(watermarked_img, watermark_text, (text_x, text_y), font, font_scale, font_color, font_thickness, cv2.LINE_AA)

    return img, watermarked_img
```

```python
img = cv2.imread("../contents/castara.jpeg")

watermark_text = "Nice Place"

original_image, watermarked_image = add_watermark(img, watermark_text)

show(original_image, "Original")
show(watermarked_image, "Watermarked Image")
```
