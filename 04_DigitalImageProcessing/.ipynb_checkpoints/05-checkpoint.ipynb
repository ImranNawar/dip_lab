```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## Image denoising using spatial mean filters.

```python
img = cv2.imread('../contents/circuit.jpg', cv2.IMREAD_GRAYSCALE)
original_image = cv2.medianBlur(img, 5)

# Add Gaussian noise with variance 400 to create the noisy image
noise = np.random.normal(0, 20, original_image.shape)
noisy_image_gaussian = original_image + noise.astype(np.uint8)

# Add pepper noise with probability 0.1
pepper_noise = np.random.rand(*original_image.shape)
noisy_image_pepper = original_image.copy()
noisy_image_pepper[pepper_noise < 0.1] = 0

# Add salt noise with probability 0.1
salt_noise = np.random.rand(*original_image.shape)
noisy_image_salt = original_image.copy()
noisy_image_salt[salt_noise < 0.1] = 255

# Apply arithmetic mean filter
arithmetic_mean_gaussian = cv2.blur(noisy_image_gaussian, (3, 3))
arithmetic_mean_pepper = cv2.blur(noisy_image_pepper, (3, 3))
arithmetic_mean_salt = cv2.blur(noisy_image_salt, (3, 3))

# Apply geometric mean filter
geometric_mean_gaussian = cv2.pow(noisy_image_gaussian.astype(np.float32), 0.1)
geometric_mean_pepper = cv2.pow(noisy_image_pepper.astype(np.float32), 0.1)
geometric_mean_salt = cv2.pow(noisy_image_salt.astype(np.float32), 0.1)

# Display the results
plt.figure(figsize=(15, 10))

plt.subplot(331), plt.imshow(original_image, cmap='gray'), plt.title('Original Image')
plt.subplot(332), plt.imshow(noisy_image_gaussian, cmap='gray'), plt.title('Noisy Image (Gaussian)')
plt.subplot(333), plt.imshow(noisy_image_pepper, cmap='gray'), plt.title('Noisy Image (Pepper)')
plt.subplot(334), plt.imshow(noisy_image_salt, cmap='gray'), plt.title('Noisy Image (Salt)')

plt.subplot(335), plt.imshow(arithmetic_mean_gaussian, cmap='gray'), plt.title('Arithmetic Mean (Gaussian)')
plt.subplot(336), plt.imshow(arithmetic_mean_pepper, cmap='gray'), plt.title('Arithmetic Mean (Pepper)')
plt.subplot(337), plt.imshow(arithmetic_mean_salt, cmap='gray'), plt.title('Arithmetic Mean (Salt)')

plt.subplot(338), plt.imshow(geometric_mean_gaussian, cmap='gray'), plt.title('Geometric Mean (Gaussian)')
plt.subplot(339), plt.imshow(geometric_mean_pepper, cmap='gray'), plt.title('Geometric Mean (Pepper)')

plt.tight_layout()
plt.show()
```

## Image denoising (interference reduction) using notch filtering

```python
noisy_image = cv2.imread('../contents/rose.tif', cv2.IMREAD_GRAYSCALE)

# Perform Fourier Transform to convert the image to frequency domain
f_transform = np.fft.fft2(noisy_image)
f_transform_shifted = np.fft.fftshift(f_transform)

# Define the coordinates of the noise frequencies to be removed (notch filter)
center_x, center_y = noisy_image.shape[1] // 2, noisy_image.shape[0] // 2
radius = 30

# Create a mask to zero-out the frequencies in the notch filter
mask = np.ones_like(noisy_image)
rows, cols = np.ogrid[:noisy_image.shape[0], :noisy_image.shape[1]]
mask[(rows - center_y)**2 + (cols - center_x)**2 < radius**2] = 0

# Apply the mask to the frequency domain image
f_transform_shifted_filtered = f_transform_shifted * mask

# Perform Inverse Fourier Transform to get the denoised image
denoised_image = np.abs(np.fft.ifft2(np.fft.ifftshift(f_transform_shifted_filtered)))

# Display the results
plt.figure(figsize=(10, 7))

plt.subplot(131), plt.imshow(noisy_image, cmap='gray'), plt.title('Noisy Image'), plt.axis('off')
plt.subplot(132), plt.imshow(np.log(1 + np.abs(f_transform_shifted)), cmap='gray'), plt.title('Frequency Domain'), plt.axis('off')
plt.subplot(133), plt.imshow(denoised_image, cmap='gray'), plt.title('Denoised Image'), plt.axis('off')

plt.tight_layout()
plt.show()
```

## Obtaining backprojected images from sinograms. 

```python
def backproject(sinogram):
    num_projections, num_angles = sinogram.shape
    num_pixels = num_projections

    reconstructed_image = np.zeros((num_pixels, num_pixels), dtype=float)

    for angle_index in range(num_angles):
        projection = sinogram[:, angle_index]
        projection_matrix = np.tile(projection, (num_pixels, 1))

        # Accumulate projections using backprojection
        reconstructed_image += np.transpose(projection_matrix)

    return reconstructed_image

# Example sinogram (replace this with your actual sinogram data)
sinogram_data = np.random.rand(180, 360)

# Perform backprojection
reconstructed_image = backproject(sinogram_data)

# Display the results
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(sinogram_data, cmap='gray', extent=(0, 360, 0, 180), aspect='auto')
plt.title('Sinogram')
plt.xlabel('Projection Angle (degrees)')
plt.ylabel('Projection Index')

plt.subplot(1, 2, 2)
plt.imshow(reconstructed_image, cmap='gray', extent=(0, 180, 0, 180), aspect='auto')
plt.title('Reconstructed Image')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()
```
