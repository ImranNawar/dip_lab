```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(image, title=None):
    plt.imshow(image, 'gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
```

## Segmentation by region growing

```python
def region_growing(image, seed):
    # Parameters
    neighbors = [(1, 0), (-1, 0), (0, 1), (0, -1)]  # 4-connectivity
    # Output mask and visited matrix
    h, w = image.shape[:2]
    segmented = np.zeros_like(image, dtype=np.uint8)
    visited = np.zeros_like(image, dtype=bool)
    # Region growing
    stack = [seed]
    while stack:
        x, y = stack.pop()
        if 0 <= x < h and 0 <= y < w and not visited[x, y]:
            # Check similarity criterion (you can modify this based on your application)
            if abs(int(image[x, y]) - int(image[seed])) < 20:
                segmented[x, y] = image[x, y]
                visited[x, y] = True
                stack.extend((x + dx, y + dy) for dx, dy in neighbors)
    return segmented

# Load an image
image_path = '../contents/brain.png'
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Choose a seed point (you can modify this based on your application)
seed_point = (120, 120)

# Apply region growing segmentation
segmented_image = region_growing(original_image, seed_point)

# Display the results
plt.imshow( original_image,'gray' ); plt.show()
plt.imshow(segmented_image, 'gray'); plt.show()
```

## Segmentation by region splitting and merging

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def region_splitting_and_merging(image, threshold):
    h, w = image.shape[:2]

    # Check the homogeneity criterion
    def is_homogeneous(region):
        if region.size == 0:
            return True  # If the region is empty, consider it homogeneous
        return np.max(region) - np.min(region) < threshold

    # Recursive region splitting
    def split_region(region, start_row, start_col, height, width):
        if height == 1 and width == 1:
            return [(start_row, start_col)]  # Return the pixel coordinates of the region

        half_height, half_width = height // 2, width // 2

        # Divide the region into four quadrants
        top_left = region[start_row:start_row + half_height, start_col:start_col + half_width]
        top_right = region[start_row:start_row + half_height, start_col + half_width:start_col + width]
        bottom_left = region[start_row + half_height:start_row + height, start_col:start_col + half_width]
        bottom_right = region[start_row + half_height:start_row + height, start_col + half_width:start_col + width]

        # Check homogeneity and split recursively if not homogeneous
        if not is_homogeneous(top_left):
            coordinates = split_region(region, start_row, start_col, half_height, half_width)
        else:
            coordinates = [(i, j) for i in range(start_row, start_row + half_height)
                                     for j in range(start_col, start_col + half_width)]

        if not is_homogeneous(top_right):
            coordinates += split_region(region, start_row, start_col + half_width, half_height, width - half_width)
        else:
            coordinates += [(i, j) for i in range(start_row, start_row + half_height)
                                      for j in range(start_col + half_width, start_col + width)]

        if not is_homogeneous(bottom_left):
            coordinates += split_region(region, start_row + half_height, start_col, height - half_height, half_width)
        else:
            coordinates += [(i, j) for i in range(start_row + half_height, start_row + height)
                                      for j in range(start_col, start_col + half_width)]

        if not is_homogeneous(bottom_right):
            coordinates += split_region(region, start_row + half_height, start_col + half_width, height - half_height, width - half_width)
        else:
            coordinates += [(i, j) for i in range(start_row + half_height, start_row + height)
                                      for j in range(start_col + half_width, start_col + width)]

        return coordinates

    # Create a copy of the input image to store the segmented result
    segmented = np.copy(image)

    # Start the region splitting process
    coordinates = split_region(segmented, 0, 0, h, w)

    # Merge adjacent homogeneous regions
    for coord in coordinates:
        region_values = image[coord[0]:coord[0]+1, coord[1]:coord[1]+1]
        mean_value = np.mean(region_values)
        segmented[coord[0]:coord[0]+1, coord[1]:coord[1]+1] = mean_value

    return segmented.astype(np.uint8)

# Load an image
image_path = '../contents/Fig0621(a)(weld-original).tif'
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Set the threshold for homogeneity (you can modify this based on your application)
threshold_value = 20

# Apply region splitting and merging segmentation
segmented_image = region_splitting_and_merging(original_image, threshold_value)

plt.imshow(original_image, cmap='gray')
plt.title('Original Image')
plt.show()

plt.imshow(segmented_image, cmap='gray')
plt.title('Segmented Image')
plt.show()
```
