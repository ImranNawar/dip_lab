```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

##  basic Fourier Transform analysis 

```python
image = cv2.imread('../contents/rose.tif', cv2.IMREAD_GRAYSCALE)

# Apply Fourier Transform
f_transform = np.fft.fft2(image)
f_transform_shifted = np.fft.fftshift(f_transform)

# Compute magnitude spectrum
magnitude_spectrum = np.log(np.abs(f_transform_shifted) + 1)

plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.show()

plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('Magnitude Spectrum (Fourier Transform)')
plt.colorbar()
plt.show()
```

## Obtaining the Fourier Transform of a simple continous function

```python
# Define the continuous function (sinusoidal signal)
def continuous_function(t):
    return np.sin(2 * np.pi * t)

# Generate time values
t = np.linspace(0, 1, 1000, endpoint=False)  # 1 second, 1000 samples

# Generate the continuous signal
signal = continuous_function(t)

# Compute the Fourier Transform
f_transform = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(t), t[1] - t[0])

# Original signal
plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.title('Original Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

# Plot the magnitude spectrum (Fourier Transform)
plt.subplot(2, 1, 2)
plt.plot(frequencies, np.abs(f_transform))
plt.title('Magnitude Spectrum (Fourier Transform)')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
