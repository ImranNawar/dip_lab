```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(image, title=None):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

## Using erosion to remove image component

```python
img = cv2.imread('../contents/wirebond.png', 0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernal = np.ones((12, 12), np.uint8)

img_erode = cv2.erode(img, kernal, iterations=1)

plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img_erode)
plt.title('Original')
plt.axis('off')

plt.show()
```

## Using dilation to repair broken characters in an image.

```python
img = cv2.imread('../contents/broken_characters.png', 0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernal = np.ones((3,3), np.uint8)

img_dilate = cv2.dilate(img, kernal, iterations=1)

plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img_dilate)
plt.title("Dilation")
plt.axis('off')

plt.show()
```

## Using opening and closing for morphological filtering.

```python
img = cv2.imread('../contents/fingerprint.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
_, mask = cv2.threshold(img, 220, 255, cv2.THRESH_BINARY_INV)

kernal = np.ones((2, 2), np.uint8)

img_Opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernal)

img_Closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernal)

plt.subplot(1,3,1)
plt.imshow(img)
plt.title('Original')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(img_Opening)
plt.title('Opening')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(img_Closing)
plt.title('Closing')
plt.axis('off')

plt.show()
```

## Boundary Extraction

```python
img = cv2.imread('../contents/boundary.png', 0)

kernal = np.ones((3, 3), np.uint8)

img_erode = cv2.erode(img, kernal, iterations=1)

boundary = img - img_erode

# Display the images
show(img, 'Original Image')
show(boundary, 'Boundary (A - A eroded by B)')
```
